```python
import matplotlib.pyplot as plt
from datetime import datetime

# Data from the table
dates = [
    "31/12/2024", "02/01/2025", "02/01/2025", "03/01/2025", "03/01/2025",
    "03/01/2025", "05/01/2025", "05/01/2025", "05/01/2025", "11/01/2025",
    "12/01/2025", "13/01/2025", "13/01/2025", "13/01/2025", "13/01/2025",
    "13/01/2025", "13/01/2025", "13/01/2025", "13/01/2025", "13/01/2025",
    "13/01/2025"
]

public_scores = [
    0.232380952, 0.139999999, 0.438709677, 0, 0.03649635, 0.287841191,
    0.248366013, 0.259030837, 0.248366013, 0.040797824, 0.010752688,
    0.070934256, 0.290456431, 0.3671875, 0.448717948, 0.413114754,
    0.456591639, 0.333333333, 0.406666666, 0.445859872, 0.402684563
]

# Convert dates to cumulative days for plotting
start_date = datetime.strptime(dates[0], "%d/%m/%Y")
days = [(datetime.strptime(date, "%d/%m/%Y") - start_date).days for date in dates]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(days, public_scores, marker='o', label="Public Score")

# Adding labels and title
plt.title("Submission Performance Over Time", fontsize=14)
plt.xlabel("Days Since First Submission", fontsize=12)
plt.ylabel("Public Score", fontsize=12)
plt.xticks(days, labels=dates, rotation=45, fontsize=8)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Techniques and corresponding average public scores
techniques = [
    "Starter NN", "MinMax Scaling", "One-Hot Encoding", "Oversampling",
    "Additional Features", "Outlier Handling", "Early Stopping", 
    "Learning Rate Scheduling", "Leaky ReLU", "L2 Regularization",
    "Batch Normalization", "Increased Epochs", "Dropout Rates"
]

average_public_scores = [
    0.232, 0.140, 0.439, 0.036, 0.288, 0.248, 0.045,
    0.011, 0.071, 0.290, 0.367, 0.448, 0.399
]

# Plotting the bar chart
plt.figure(figsize=(12, 6))
bars = plt.bar(techniques, average_public_scores, color='skyblue', edgecolor='black')

# Adding value annotations on bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.3f}', 
             ha='center', va='bottom', fontsize=10)

# Adding labels and title
plt.title("Effectiveness of Techniques Based on Public Scores", fontsize=14)
plt.xlabel("Techniques", fontsize=12)
plt.ylabel("Average Public Score", fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
